```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

def funkcja(x):
  return x*x+5


x = np.linspace(-1, 1)

y = funkcja(x)
plt.figure(1)
plt.plot(x,y, label="funkcja")
plt.title("x^2+5")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()


```
